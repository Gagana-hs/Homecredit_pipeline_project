```python
from pyspark.sql.functions import col, when, count

def count_nulls(df):
    """
    Returns null count for every column in a DataFrame.
    """

    return df.select([
        count(when(col(c).isNull(), c)).alias(c)
        for c in df.columns
    ])
```

```python
from pyspark.sql.types import (
    StringType,
    IntegerType,
    LongType,
    FloatType,
    DoubleType,
    DecimalType
)

def clean_nulls(df, key_columns=None):

    # Drop rows with null primary keys
    if key_columns:
        existing = [c for c in key_columns if c in df.columns]

        if existing:
            df = df.dropna(subset=existing)

    # Fill numeric columns with median
    numeric_fill = {}

    numeric_types = (
        IntegerType,
        LongType,
        FloatType,
        DoubleType,
        DecimalType
    )

    for field in df.schema.fields:

        if isinstance(field.dataType, numeric_types):

            median = df.approxQuantile(
                field.name,
                [0.5],
                0.01
            )[0]

            numeric_fill[field.name] = median

    if numeric_fill:
        df = df.fillna(numeric_fill)

    # Fill string columns
    string_fill = {}

    for field in df.schema.fields:

        if isinstance(field.dataType, StringType):

            string_fill[field.name] = "Unknown"

    if string_fill:
        df = df.fillna(string_fill)

    return df
```

```python
def process_tables(table_list, schema_name):

    for table in table_list:

        print("=" * 70)
        print(f"Processing : {table}")

        df = spark.table(f"{schema_name}.{table}")

        print("\nNull values BEFORE cleaning")

        display(count_nulls(df))

        df = clean_nulls(
            df,
            key_columns=["SK_ID_CURR"]
        )

        print("\nNull values AFTER cleaning")

        display(count_nulls(df))

        (
            df.write
              .format("delta")
              .mode("overwrite")
              .option("overwriteSchema", "true")
              .saveAsTable(f"{schema_name}.{table}_silver")
        )

        print(f"{table}_silver created successfully")
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import (
    IntegerType,
    DoubleType,
    DateType
)

def fix_data_types(df):

    # Integer columns
    int_columns = [
        "SK_ID_CURR",
        "SK_ID_BUREAU",
        "SK_ID_PREV",
        "TARGET"
    ]

    # Double columns
    double_columns = [
        "AMT_INCOME_TOTAL",
        "AMT_CREDIT",
        "AMT_ANNUITY",
        "AMT_GOODS_PRICE"
    ]

    # Date columns (if present)
    date_columns = [
        "DATE_DECISION"
    ]

    # Convert Integer columns
    for c in int_columns:
        if c in df.columns:
            df = df.withColumn(c, col(c).cast(IntegerType()))

    # Convert Double columns
    for c in double_columns:
        if c in df.columns:
            df = df.withColumn(c, col(c).cast(DoubleType()))

    # Convert Date columns
    for c in date_columns:
        if c in df.columns:
            df = df.withColumn(c, col(c).cast(DateType()))

    return df
```

```python
from pyspark.sql.functions import col

def apply_business_rules(df):

    # Credit amount must be positive
    if "AMT_CREDIT" in df.columns:
        df = df.filter(col("AMT_CREDIT") > 0)

    # Income must be positive
    if "AMT_INCOME_TOTAL" in df.columns:
        df = df.filter(col("AMT_INCOME_TOTAL") > 0)

    # Goods price cannot be negative
    if "AMT_GOODS_PRICE" in df.columns:
        df = df.filter(col("AMT_GOODS_PRICE") >= 0)

    # Annuity cannot be negative
    if "AMT_ANNUITY" in df.columns:
        df = df.filter(col("AMT_ANNUITY") >= 0)

    # Target must be either 0 or 1
    if "TARGET" in df.columns:
        df = df.filter(col("TARGET").isin([0,1]))

    return df
```

```python
def save_silver_table(df, schema_name, table_name):

    (
        df.write
          .format("delta")
          .mode("overwrite")
          .option("overwriteSchema","true")
          .saveAsTable(f"{schema_name}.{table_name}_silver")
    )

    print(f"{table_name}_silver created successfully")
```

```python
def silver_pipeline(table_name, schema_name):

    print("="*70)
    print(f"Processing {table_name}")

    # Read Bronze table
    df = spark.table(f"{schema_name}.{table_name}")

    print("Rows before cleaning :", df.count())

    # Null handling
    df = clean_nulls(
        df,
        key_columns=["SK_ID_CURR"]
    )

    # Fix data types
    df = fix_data_types(df)

    # Apply business rules
    df = apply_business_rules(df)

    print("Rows after cleaning :", df.count())

    # Save Silver
    save_silver_table(
        df,
        schema_name,
        table_name
    )

    print(f"{table_name} completed")
```

```python
tables = [

    "application_train",
    "application_test",
    "bureau",
    "bureau_balance",
    "credit_card_balance",
    "installments_payments",
    "pos_cash_balance",
    "previous_application"

]

for table in tables:

    silver_pipeline(
        table,
        "insurance.homecredit"
    )
```
